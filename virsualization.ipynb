```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./log.csv')
df[:].plot(x="generation", y=["avarage fitness", "global best fitness", "local best fitness"])
df[:].plot(x="generation", y=["mutation rate"])
df[-1:]
```

# 1. Mutation rate comparision

```python

df_mut_low = pd.read_csv('./log_mutation_0.001.csv')
df_mut_high = pd.read_csv('./log_mutation_0.005.csv')
df_mut_very_high = pd.read_csv('./log_mutation_0.01.csv')

fig_1, (ax1, ax2) = plt.subplots(1, 2)
fig_1.set_figwidth(15)
ax1.plot(df_mut_low["generation"], df_mut_low[["avarage fitness", "global best fitness", "local best fitness"]])
ax2.plot(df_mut_high["generation"], df_mut_high[["avarage fitness", "global best fitness", "local best fitness"]])

ax1.title.set_text("Low mutation rate (0.001)")
ax2.title.set_text("High mutation rate (0.005)")
```
